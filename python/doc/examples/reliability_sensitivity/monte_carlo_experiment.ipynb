# Create a Monte Carlo design of experiments

In this basic example we are going to create a MonteCarlo probabilistic design experiment.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Define underlying distribution, size
distribution = ot.Normal(2)
size = 50
```

```python
# Create the design
myPlane = ot.MonteCarloExperiment(distribution, size)
sample = myPlane.generate()
```

```python
# Plot the design
graph = ot.Graph("MC design", "x1", "x2", True, "")
cloud = ot.Cloud(sample, "blue", "fsquare", "")
graph.add(cloud)
graph
```
