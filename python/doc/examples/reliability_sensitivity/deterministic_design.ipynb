# Deterministic design of experiments

In this example we are going to expose the available deterministic design of experiments that allow for example
to determine the variation range of an output variable of interest.

Four types of deterministic design of experiments are available:

- Axial
- Factorial
- Composite
- Box

Each type of deterministic design is discretized differently according to a number of levels.

The generated samples lies in the unit cube $(0,1)^d$ and can be scaled then translated to cover the desired interval.


```python
from __future__ import print_function
import openturns as ot
import math as m
```

```python
# Axial design
levels = [1.0, 1.5, 3.0]
experiment = ot.Axial(2, levels)
sample = experiment.generate()
ot.Cloud(sample)
```

```python
# scale & translate to get desired location
sample *= 2.0
sample += [5.0, 8.0]
ot.Cloud(sample)
```

```python
# Factorial design
experiment = ot.Factorial(2, levels)
sample = experiment.generate()
sample *= 2.0
sample += [5.0, 8.0]
ot.Cloud(sample)
```

```python
# Composite design
experiment = ot.Composite(2, levels)
sample = experiment.generate()
sample *= 2.0
sample += [5.0, 8.0]
ot.Cloud(sample)
```

```python
# Grid design
levels = [3, 4]
experiment = ot.Box(levels)
sample = experiment.generate()
sample *= 2.0
sample += [5.0, 8.0]
ot.Cloud(sample)
```
