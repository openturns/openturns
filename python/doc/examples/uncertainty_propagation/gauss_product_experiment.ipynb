# Create a Gauss product design of experiments

In this basic example we are going to create a deterministic weighted design experiment using Gauss product.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Define the underlying distribution, degrees
distribution = ot.ComposedDistribution([ot.Exponential(), ot.Triangular(-1.0, -0.5, 1.0)])
marginalDegrees = [15, 8]
```

```python
# Create the design
myPlane = ot.GaussProductExperiment(distribution, marginalDegrees)
sample = myPlane.generate()
```

```python
# Plot the design
graph = ot.Graph("GP design", "x1", "x2", True, "")
cloud = ot.Cloud(sample, "blue", "fsquare", "")
graph.add(cloud)
graph
```
