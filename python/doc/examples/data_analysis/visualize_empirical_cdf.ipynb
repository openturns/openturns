# Visualize empirical CDF

In this example we are going to draw the empirical CDF of an unidimensional sample.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Create data to visualize
size = 100
normal = ot.Normal(1)
sample = normal.getSample(size)
```

```python
# Draw empirial CDF
xmin = sample.getMin()[0] - 1.0
xmax = sample.getMax()[0] + 1.0
ot.VisualTest.DrawEmpiricalCDF(sample, xmin, xmax)
```
