# Estimate an integral

In this basic example we are going to evaluate an integral of the form.

$$I_f = \int_{a}^{b}\, f(x)\mathrm{d}{x}$$

with the iterated quadrature algorithm.

```python
from __future__ import print_function
import openturns as ot
import math as m
```

```python
# define the integrand and the bounds
a = -m.pi
b = m.pi
f = ot.SymbolicFunction(['x', 'y'], ['1+cos(x)*sin(y)'])
```

```python
# compute the integral value
algo = ot.ClenshawCurtis([20,20])
I2,nodes = algo.integrateWithNodes(f,ot.Interval([a,a],[b,b]))
print(I2)
algo2 = ot.GaussLegendre([20,20])
I,node = algo2.integrateWithNodes(f,ot.Interval([a,a],[b,b]))
print(I)
```

```python
# Draw the graph of the integrand and the bounds
g = ot.Graph('Integration nodes', 'x', 'y', True, 'topright')
g.add(f.draw([a,a],[b,b]))

cloud = ot.Cloud(nodes)
g.add(cloud)
g
```
