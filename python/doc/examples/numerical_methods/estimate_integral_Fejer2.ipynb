# Estimate an integral (with Gauss Legendre and Fejer2)

In this basic example we are going to evaluate an integral of the form.

$$I_f = \int_{a}^{b}\, \int_{c}^{d}f(x,y)\mathrm{d}{x}\mathrm{d}{y}$$

with the iterated quadrature algorithm.

```python
from __future__ import print_function
import openturns as ot
import math as m
```

```python
# define the integrand and the bounds
a = -m.pi
b = m.pi/2
c = -m.pi
d = m.pi/2
f = ot.SymbolicFunction(['x', 'y'], ['1+cos(x)*sin(y)'])
#l = [ot.SymbolicFunction(['x'], [' 2+cos(x)'])]
#u = [ot.SymbolicFunction(['x'], ['-2-cos(x)'])]
```

```python
# Draw the graph of the integrand and the bounds
g = ot.Graph('Integration nodes', 'x', 'y', True, 'topright')
g.add(f.draw([a,c],[b,d]))
g
```

```python
# compute the integral value with Gauss Legendre
algo = ot.GaussLegendre([10,10])
value, adaptedNodes = algo.integrateWithNodes(f, ot.Interval([a,a],[b,b]))
print(value,adaptedNodes)
```

```python
# Draw the graph of the integrand and the bounds
g = ot.Graph('Integration nodes', 'x', 'y', True, 'topright')
g.add(f.draw([a,c],[b,d]))
cloud = ot.Cloud(adaptedNodes)
cloud.setColor("black")
cloud.setPointStyle('plus')
g.add(cloud)
g
```

```python
# compute the integral value with Fejer2
algoF2 = ot.Fejer2([10,10])
valueF2, adaptedNodesF2 = algoF2.integrateWithNodes(f, ot.Interval([a,a],[b,b]))
print(valueF2,adaptedNodesF2)
```

```python
# Draw the graph of the integrand and the bounds
g = ot.Graph('Integration nodes', 'x', 'y', True, 'topright')
g.add(f.draw([a,c],[b,d]))
cloud = ot.Cloud(adaptedNodesF2)
cloud.setColor("black")
cloud.setPointStyle('plus')
g.add(cloud)
g
```
